```
import numpy as np 
import matplotlib.pyplot as plt 
 
x = np.linspace(0,10,1000)
y = np.sin(x)
z = np.cos(x**2)
plt.plot(x,y, label="$sin(x)$", color="red")
plt.plot(x,z,"b--",label="$cos(x^2)$",color="green")
plt.xlabel("Time(s)")
plt.xlabel("Volt")
plt.title("sin  cos")
plt.legend()
plt.show()
plt.savefig('plot,jpg',dpi=120)
```
